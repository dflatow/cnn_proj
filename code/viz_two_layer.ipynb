```python
import pandas as pd
import numpy as np
import cPickle as pickle
import os
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (15.0, 15.0) # set default size of plots
```

```python
filename = 'two_layer_nn_20150207_00.csv'
df = pd.read_csv(filename, sep='\t')
df.sort(columns=['hidden_size', 'reg', 'batch_size', 'learning_rate'], inplace=True)
```

```python
print pd.pivot_table(df, 'best_val', 'learning_rate', 'batch_size', aggfunc=np.mean)
```

```python
var = 'hidden_size'
others = ['hidden_size', 'reg', 'batch_size', 'learning_rate']
others = ['hidden_size', 'learning_rate', 'batch_size', 'reg']
others.remove(var)

df = df[df['batch_size'] == 600]

df.sort(columns=['hidden_size', 'reg', 'batch_size', 'learning_rate'], inplace=True)
g = df.groupby(others)
```

```python
rows, cols = len(g)/3, 3
f, axes = plt.subplots(rows, cols, sharex='col', sharey='row')
f.subplots_adjust(hspace=1.0)


c = 0
for i, x in g:
  
    axes[c%rows, c/rows].scatter(x[var], x['best_val'])
    axes[c%rows, c/rows].set_title(x[others].iloc[0].to_string())
    c += 1
```

```python
best_model_file = 'best_two_layer_nn.p'
with open(best_model_file) as f:
    best_model, best_val_acc = pickle.load(f)
    print 'Standing val record: %f' % best_val_acc
        
from cs231n.vis_utils import visualize_grid

# Visualize the weights of the network

def show_net_weights(model):
    plt.imshow(visualize_grid(model['W1'].T.reshape(-1, 32, 32, 3), 
                              padding=3).astype('uint8'))
    plt.gca().axis('off')
    plt.show()

show_net_weights(best_model)
```
